```python
import nltk
```

```python
sentence = """At eight o'clock on Thursday morning Arthur didn't feel very good."""
```

**Demonstrating basic functionality of the NLTK Tokenizer.**

```python
tokens = nltk.word_tokenize(sentence)
```

```python
tokens
```

**Here, the tokens generated by the Tokenizer are tagged by keeping track of their positions in the sentence. Again, all of this is to just try out the basic functionality of the NLTK Library.**

```python
tagged = nltk.pos_tag(tokens)
```

```python
tagged
```

```python
entities = nltk.chunk.ne_chunk(tagged)
```

```python
entities
```

```python
from nltk.corpus import wordnet as wn
```

**Synsets in WordNet basically tell us about all the Lemmas that are similar to the given Lemma. They work through establishing the relations among various concepts available in the Corpus.**

```python
sen = wn.synsets('game')
```

```python
sen
```

**Extracting definition for the given lemma, "game".**

```python
sen[0].definition()
```

**Here we go on exploring the entire synset data available in WordNet.**

```python
for sen in wn.all_synsets():
    print(sen)
    print(sen.definition())
    break
```

```python
lemmas = set(i for i in wn.all_synsets())
```

```python
words = list(x for x in wn.words())
```

**Counting the total Lemmas and Words available in WordNet.**

```python
len(lemmas)
```

```python
len(words)
```

```python
lemma_names = [ss.lemma_names() for ss in wn.all_synsets()]
```

```python
len(lemma_names)
```

```python
def gen(words):
    """
    Simple generator that takes any iterable object as input.
    """
    for _ in words:
        yield _
```

**Extracting and parsing through the definitions available for every single concept in WordNet.**

```python
i = 0
for _ in gen(lemmas):
    print(str(_) + " : " + _.definition())
    i += 1
    if i == 10:
        break
```
